# Résolution

On cherche le point le plus haut d'une montagne représentée par une fonction de deux variables $x$ et $y$ les abscisses et ordonnées sur la carte et $z$ la côte (ou altitude). On cherche donc le point $(x,y)$ tel que $z = f(x,y)$ soit maximal.

$ z = f(x,y) = xye^{-(x,y)}$

## Etude des points critiques de la fonction $f$
Les point critiques s'obtiennent lorsque le gradient de $f$ est nul. Soit $\overrightarrow{\nabla} f = \left( \frac{\partial f}{\partial x} , \frac{\partial f}{\partial y}\right) = (0,0)$.

### Calcul des dérivées partielles

$
\begin{aligned}
\frac{\partial f(x,y)}{\partial x} = y e^{-(x,y)} + x y e^{-(x,y)} \\[10pt]
\frac{\partial f(x,y)}{\partial y} = x e^{-(x,y)} + x y e^{-(x,y)}
\end{aligned}
$

Donc $\text{grad}\ f = (e^{-(x,y)}(y-xy),\ e^{-(x,y)} (x-xy))$

On cherche donc à résoudre le système d'équations suivant :

$\left\{
\begin{aligned}
e^{-(x,y)}(y-xy) = 0 \\
e^{-(x,y)} (x-xy) = 0
\end{aligned}
\right.$

Car la fonction exponencielle est strictement positive.

Soit quand 
$\left\{
\begin{aligned}
y - xy = 0 \\
x - xy = 0
\end{aligned}
\right.
\Leftrightarrow
\left\{
\begin{aligned}
y(1 - x) = 0 \\
x(1 - y) = 0
\end{aligned}
\right.
\Leftrightarrow
\left\{
\begin{aligned}
y = 0 \text{ ou } x = 1 \\
x = 0 \text{ ou } y = 1
\end{aligned}
\right.$

Donc on a deux points critiques en $(0,0)$ et en $(1,1)$.

### Déterminer la nature des points critiques

En utilisant la méthode du déterminant, on peut déterminer si le point critique est un minimum local ou un maximum local.

Le déterminant est donné par la matrice hessienne de la fonction $f(x,y)$.

$ H = \left[ \begin{array}{cc}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\[10pt]
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
\end{array} \right]$

Le déterminant est donné $\text{Det}(H) = \frac{\partial^2 f}{\partial x^2} \times \frac{\partial^2 f}{\partial y^2} - \left( \frac{\partial^2 f}{\partial x \partial y} \right)^2$

Soit dans notre cas :

$\begin{aligned}
\frac{\partial^2 f}{\partial x^2} &= xye^{-(x,y)} - 2ye^{-(x,y)} \\[10pt]
\frac{\partial^2 f}{\partial x \partial y} &= e^{-(x,y)} - ye^{-(x,y)} - xe^{-(x,y)} + xye^{-(x,y)} \\[10pt]
\frac{\partial^2 f}{\partial y \partial x} &= e^{-(x,y)} - xe^{-(x,y)} - ye^{-(x,y)} + xye^{-(x,y)} \\[10pt]
\frac{\partial^2 f}{\partial y^2} &= xye^{-(x,y)} - 2xe^{-(x,y)}
\end{aligned}$

#### Point critique en $(1,1)$

En évaluant la dérivée partielle en $(1,1)$, on obtient :

$\begin{aligned}
\frac{\partial^2 f}{\partial x^2} (1,1) &= e^{-2} - 2e^{-2} &= -e^{-2} \approx -0.135 \\[10pt]
\frac{\partial^2 f}{\partial x \partial y} (1,1) &= e^{-2} - e^{-2} - e^{-2} + e^{-2} &= 0 \\[10pt]
\frac{\partial^2 f}{\partial y \partial x} (1,1) &= e^{-2} - e^{-2} - e^{-2} + e^{-2} &= 0 \\[10pt]
\frac{\partial^2 f}{\partial y^2} (1,1) &= e^{-2} - 2e^{-2} &= -e^{-2} \approx -0.135
\end{aligned}$

Donc 
$\begin{aligned}
\text{Det}(H) &= (-e^{-2})^2
&= e^{-4} \approx 0.018
\end{aligned}$

Puisque $\text{Det}(H) > 0$ et $\frac{\partial^2 f}{\partial x^2} (1,1) < 0$ alors le point critique est un maximum local.

#### Points critiques en $(0,0)$

En évaluant la dérivée partielle en $(0,0)$, on obtient :

$\begin{aligned}
\frac{\partial^2 f}{\partial x^2} (0,0) &= 0e^{0} - 2 \times 0 e^{0} &= 0 \\[10pt]
\frac{\partial^2 f}{\partial x \partial y} (0,0) &= e^{0} - 0e^{0} - 0e^{0} + 0e^{0} &= 1 \\[10pt]
\frac{\partial^2 f}{\partial y \partial x} (0,0) &= e^{0} - 0e^{0} - 0e^{0} + 0e^{0} &= 1 \\[10pt]
\frac{\partial^2 f}{\partial y^2} (0,0) &= 0e^{0} - 2 \times 0 e^{0} &= 0
\end{aligned}$

Donc 
$\begin{aligned}
\text{Det}(H) &= 0 \times 0 - 1^2
&= -1
\end{aligned}$

Puisque $\text{Det}(H) < 0$ alors le point critique est un point de selle. C'est à dire que la fonction $f$ passe de concave à convexe ou de convexe à concave en ce point.

> On peut donc déduire que puisque le point $(1,1)$ est le seul maximum local, il est l'unique sommet du relief défini par la fonction $f$. Et donc les coordonnées du sommet le plus haut sur la carte sont $(1,1)$.

## Calcul de l'altitude du sommet en $(1,1)$ avec mise à l'échelle

$\begin{aligned}
z &= xye^{-(x+y)}
&= e^{-2}
&\approx 0.135
\end{aligned}$

Sachant que l'échelle de la carte est 1cm pour 50 000cm soit 1cm pour 500m, on a :
$e^{-2} \times 500 \approx 67.67m$ d'altitude.

> Pour conclure, par rapport à leur point de départ, les deux randonneurs parcoureront un dénivelé de 67.67m depuis le point de départ à l'altitude de 0m ($z=0$) pour se rendre au sommet le plus haut de la carte en $(1,1)$.

## Illustration
Voici la représetation graphique de la fonction $f$ avec le sommet le plus haut en $(1,1)$ au point $A$ sur la figure.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Définition de la fonction f(x,y)
def f(x, y):
    return x * y * np.exp(-1 * (x + y))

# Points A et B
A = (1, 1, math.exp(-2))
B = (1, 1, 0)

# Coordonnées x, y et z pour la fonction f(x,y)
x = y = np.arange(0.0, 2.0, 0.05)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Affichage de la fonction f(x,y) en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Change transparence map
ax.plot_surface(X, Y, Z, alpha=0.5)

# Affichage des points A et B
ax.scatter(A[0], A[1], A[2], c='r', marker='o')
# Add big annotation
ax.text(A[0], A[1], A[2], 'A', size=20, zorder=1, color='k')
ax.scatter(B[0], B[1], B[2], c='r', marker='o')

# Affichage du segment entre A et B
ax.plot([A[0], B[0]], [A[1], B[1]], [A[2], B[2]], c='r')

# Paramètres de l'affichage
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
# Change fig size
fig.set_size_inches(10, 10)
plt.show()
```
